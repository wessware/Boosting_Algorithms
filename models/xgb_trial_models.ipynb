```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as mse
import xgboost as xgb
```

```python
df_xg = pd.read_csv('../datasets/clean_data_loan_modelling.csv')
```

```python
df_xg
```

```python
x = df_xg.drop(columns='Loan_Status')
y = df_xg['Loan_Status']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
xgb_model = xgb.XGBClassifier(
    #colsample_bytree=0.8,
    scale_pos_weight=1,
    max_depth=15, 
    n_estimators=1000,
    objective='reg:logistic',
    booster='gbtree', 
    n_jobs=-1,
    random_state=0,
    #num_class=2,
    learning_rate=0.1,
    use_label_encoder=False
)
```

```python
xgb_model.fit(x_train, y_train)
```

```python
xg_preds = xgb_model.predict(x_test)
```

```python
xgb_rmse = np.sqrt(mse(y_test, xg_preds))
```

```python
xgb_rmse
```
